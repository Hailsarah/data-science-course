```python
#read csv file as var df_1 and df_2
df_1 = pd.read_csv('01 telecom_revenue.csv')
df_2= pd.read_csv('02 telecom_usage.csv')
```

# Filter dataset 01 telecom_revenue.csv dengan multiple kondisi
 - MonthlyRevenue > 10
 - Occupation terdiri dari Professional, Student, and Crafts

```python
df_1 =  df_1 [ (df_1 ['Occupation'].isin(['Professional', 'Student',  'Crafts'])) 
                 & (df_1 ['MonthlyRevenue'] > 10) ]
df_1
```

# df_2 : Filter dataset 02 telecom_usage.csv dengan multiple kondisi
- CustomerCareCalls diawali digit genap (2, 4, 6 atau 8)
- nilai UnansweredCalls > .BlockedCalls

```python
#check type data CustomerCareCalls
df_2['CustomerCareCalls'].dtypes
```

```python
#membuat var df_2 dengan filter di atas
df_2 = df_2[df_2['CustomerCareCalls'].astype(str).str.contains('^2|^4|^6|^8')
    & df_2['UnansweredCalls'] > df_2['BlockedCalls']]
```

```python
df_merge_telecom = df_1.merge(df_2, how= 'left', on= 'CustomerID')
df_merge_telecom 
```

```python
#total customer berdasarkan occupation: crafts,professional, dan student (berdasarkan filter df_1 dan df_2)
total_customer = df_merge_telecom.groupby('Occupation').CustomerID.count()
total_customer
```

```python
#menghitung rata2 droppedcalls berdasarkan occupationnya
avg_droppedcalls = df_merge_telecom.groupby('Occupation').DroppedCalls.mean()
avg_droppedcalls
```

```python
#rata2 droppedcalls terbanyak berdasarkan occupation 
df_merge_telecom.groupby('Occupation').DroppedCalls.mean().head(1)
```
